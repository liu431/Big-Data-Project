```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
date, sentiment, popularity = [], [], []
with open("sent.txt") as f:
    for line in f:
        pair = line.split("\t") 
        date.append(pair[0][1:-1])
        sentiment.append(pair[1].split(",")[0][1:])
        popularity.append(pair[1].split(",")[1][:-2])
```

```python
len(sentiment)
```

```python
df = pd.DataFrame({"Sentiment": sentiment,
                   "Popularity": popularity}, index=date).astype("float")
```

```python
df.head()
```

```python
normdf=(df-df.mean())/df.std()
```

```python
normdf.head()
```

```python
plt.plot(normdf.index, normdf.Popularity)
```

```python
sns.set(style="white")
plt.figure(figsize=(50,40))

sns.lineplot(data=normdf, palette="tab10", linewidth=4)
plt.title("Sentiment and Popularity for Python")
plt.savefig("Sentiment and Popularity for Python.png")
```
